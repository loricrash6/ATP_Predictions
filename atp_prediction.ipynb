```python
from betting import Bets
bets = Bets(method='kelly')
```

```python
from sklearn.preprocessing import LabelEncoder

#Encode name into numeric values
name_encoder = LabelEncoder()
name_encoder.fit(pd.concat([df['p1'], df['p2']], axis=0))

df['p1_id'] = name_encoder.transform(df['p1'])
df['p2_id'] = name_encoder.transform(df['p2'])

df.drop(['p1','p2'], axis =1, inplace=True)
```

We will use a train/test split based on the year of the game. Games from 2010 to 2017 will be used as training set, and games from 2018 and 2019 will be used for testing.

```python
df_train = df[df['year']<2018]
df_test = df[df['year']>=2018]

print(f'Training samples: {df_train.shape[0]} Testing samples: {df_test.shape[0]}')
X_train = df_train.drop(['winner','uncertainty'], axis=1)
y_train = df_train["winner"]

X_test = df_test.drop(['winner','uncertainty'], axis=1)
y_test = df_test["winner"]

#X_train, X_test, y_train, y_test = train_test_split(df.drop(['winner','uncertainty'], axis=1), df['winner'], #test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score

logistic_model1 = LogisticRegressionCV(cv=5)
logistic_model1.fit(X_train, y_train)
y_hat = logistic_model1.predict(X_test)
print(f'Accuracy of Logistic regression : {accuracy_score(y_test, y_hat)}')

probs = logistic_model1.predict_proba(X_test)

roi = bets.roi(list(probs[:,0]), list(X_test['p1_odd']), list(X_test['p2_odd']), list(y_test))
print(f'ROI : {round(roi,4)*100}%')

logistic_model2 = make_pipeline(StandardScaler(), LogisticRegressionCV(cv=5))
logistic_model2.fit(X_train, y_train)
y_hat = logistic_model2.predict(X_test)
print(f'Accuracy of Logistic regression : {accuracy_score(y_test, y_hat)}')

probs = logistic_model2.predict_proba(X_test)

roi = bets.roi(list(probs[:,0]), list(X_test['p1_odd']), list(X_test['p2_odd']), list(y_test))
print(f'ROI : {round(roi,4)*100}%')
```

```python
svm1 = make_pipeline(StandardScaler(), LinearSVC())
svm1.fit(X_train, y_train)
y_hat = svm1.predict(X_test)
print(f'Accuracy of SVM : {accuracy_score(y_test, y_hat)}')

probs = svm1.predict_proba(X_test)
roi = bets.roi(list(probs[:,0]), list(X_test['p1_odd']), list(X_test['p2_odd']), list(y_test))
print(f'ROI : {round(roi,4)*100}%')
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_hat = rf.predict(X_test)
print(f'Accuracy of Random Forest : {accuracy_score(y_test, y_hat)}')

probs = rf.predict_proba(X_test)
roi = bets.roi(list(probs[:,0]), list(X_test['p1_odd']), list(X_test['p2_odd']), list(y_test))
print(f'ROI : {round(roi,4)*100}%')
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=100)
mlp.fit(X_train, y_train)
y_hat = mlp.predict(X_test)
print(f'Accuracy of MLP : {accuracy_score(y_test, y_hat)}')

probs = mlp.predict_proba(X_test)
roi = bets.roi(list(probs[:,0]), list(X_test['p1_odd']), list(X_test['p2_odd']), list(y_test))
print(f'ROI : {round(roi,4)*100}%')
```
